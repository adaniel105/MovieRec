```python
import pandas as pd
titles = pd.read_csv('titles.csv')
credits = pd.read_csv('credits.csv')
```

```python
credits.head()
```

```python
titles.head()
```

```python
df = pd.merge(left=credits, right=titles, on='id')
```

```python
df.head()
```

```python
df.groupby(['role']).agg(lambda x: x.value_counts().sum())
```

```python
features = ['name', 'title', 'type', 'release_year', 'age_certification', 'genres', 'runtime', 'seasons', 'imdb_score', 'tmdb_popularity', 'tmdb_score']

df.drop([col for col in df.columns if col not in features], axis=1, inplace=True)

df.head()
```

```python
def fill_NaN_with_None(cols):
  for col in cols:
    if df[col].isna().sum() > 0:
      df[col].fillna("None", inplace=True)

fill_NaN_with_None(['title', 'imdb_score', 'tmdb_popularity', 'tmdb_score'])
```

```python
df['seasons'].fillna("0", inplace=True)
```

```python
df = df.join(pd.get_dummies(df['age_certification'], prefix="age_cert", sparse=True))
df.drop("age_certification", axis=1,inplace=True)
df.head()
```

```python
hold = []
for i in range(len(df['genres'])):
  genre_list = df['genres'][i].split(",")
  for j in range(len(genre_list)):
    genre_str = genre_list[j].replace("[","").replace("]","").replace("'","").strip()
  hold.append(genre_str)

count = len(hold)
count
#genre = df['genres'][0].split(",")[0].replace("[","").replace("]","").replace("'","").strip()
```
