```python
import pandas as pd
titles = pd.read_csv('titles.csv')
credits = pd.read_csv('credits.csv')
```

```python
credits.head()
```

```python
titles.head()
```

```python
df = pd.merge(left=credits, right=titles, on='id')
```

```python
df.head()
```

```python
df.groupby(['role']).agg(lambda x: x.value_counts().sum())
```

```python
features = ['name', 'type', 'release_year', 'age_certification', 'genres', 'runtime', 'seasons', 'imdb_score', 'tmdb_popularity', 'tmdb_score']

df.drop([col for col in df.columns if col not in features], axis=1, inplace=True)

df.head()
```

```python
def fill_NaN_with_zero(cols):
  for col in cols:
    if df[col].isna().sum() > 0:
      df[col].fillna("0", inplace=True)

fill_NaN_with_zero(['imdb_score', 'tmdb_popularity', 'tmdb_score'])
```

```python
df['seasons'].fillna("0", inplace=True)
```

```python
df = df.join(pd.get_dummies(df['age_certification'], prefix="age_cert"))
df.drop("age_certification", axis=1,inplace=True)
df.head()
```

```python
genre_list = set()
for i in range(len(df['genres'])):
  genre_list_per_row = df['genres'][i].split(",")
  for j in range(len(genre_list_per_row)):
    genre_str = genre_list_per_row[j].replace("[","").replace("]","").replace("'","").strip()
    genre_list.add(genre_str)
genre_list.remove('')
```

```python
for genre in genre_list:
  df[f'genre_{genre}'] = 0
```

```python
for i in range(len(df['genres'])):
  genre_list_per_row_test = df['genres'][i].split(",")
  for j in range(len(genre_list_per_row_test)):
    genre_str_test = genre_list_per_row_test[j].replace("[","").replace("]","").replace("'","").strip()
    df.loc[i, f"genre_{genre_str_test}"] = 1
```

```python
df.drop(["genres", "genre_"], axis=1,inplace=True)
df.head()
```

```python
df["type"] = df["type"].map({"MOVIE" : 0, "SHOW" : 1})
df.head()
```

```python
df.drop("name", axis=1,inplace=True)
df.head()
```
